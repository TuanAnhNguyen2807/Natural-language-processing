```python
import numpy as np
import pandas as pd
import nltk 
from nltk.util import ngrams
import string
from nltk import word_tokenize
import random
import os
nltk.download('punkt')
```

```python
fpaths = []

for f in os.listdir('/content/drive/MyDrive/Kho ngu lieu 10000 cau duoc gan nhan tu loai'):
     fpaths.append('/content/drive/MyDrive/Kho ngu lieu 10000 cau duoc gan nhan tu loai/' + f)
print("All paths:", fpaths)
```

```python
# Tạo n-gram
def get_ngrams(text, n ):
    n_grams = ngrams(word_tokenize(text.lower()), n)
    return [ ' '.join(grams) for grams in n_grams]
```

```python
all_sentences = []
for path in fpaths:
    data = pd.read_table(path)
    data_sentences = data.iloc[:,-1].values
    sentences_list = [get_ngrams(data, 1) for data in data_sentences]
    for i in sentences_list:
        all_sentences.append(i)
```

```python
temp_ = []
all_sent_split = []
for i, sents in enumerate(all_sentences[0:3000]):
    for j, n in enumerate(sents):
        if n[0] not in string.punctuation:
            temp_.append(n)
        if n  == "/":
            temp = sents[j-1] + sents[j] + sents[j+1]
            temp_.append(temp)
    all_sent_split.append(temp_)
    temp_ = []
```

```python
print(all_sent_split[0:5])
print(len(all_sent_split))
```

```python
words = []
tags = []
character = ["''","'''","’","“","”","…","..."]
tagin = ['m', 'n', 'r', 'x', 's', 'b', 'y', 'c', 'e', 't', 'nu', 'nc', 'l', 'p', 'a', 'i', 'v', 'np']
for i, sent in enumerate(all_sent_split):
    for j, s in enumerate(sent):
        try:
            w, t = s.split("/")[0],  s.split("/")[1]
        except:
            continue
        else:
            if w not in string.punctuation and t not in string.punctuation and w not in character and t not in character and t in tagin:
                words.append(w)
                tags.append(t)
```

```python
print(len(words))
print(len(tags))
```

```python
df = pd.DataFrame({"Word": words, "Tag": tags})
```

```python
df.head()
```

```python
X = df.iloc[:,0].values
y = df.iloc[:,-1].values
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer 

tokenizer = Tokenizer()
tokenizer.fit_on_texts(X)
```

```python
X = tokenizer.texts_to_matrix(X,mode='binary')
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.naive_bayes import MultinomialNB 
from sklearn.metrics import classification_report

model_NB = MultinomialNB()
model_NB = model_NB.fit(X_train,y_train)

y_pred = model_NB.predict(X_test)

print(classification_report(y_test,y_pred))
```

```python
from sklearn.linear_model import LogisticRegression  
model_LR = LogisticRegression(max_iter=1000)
model_LR = model_LR.fit(X_train,y_train)

y_pred_LR = model_LR.predict(X_test)

print(classification_report(y_test,y_pred_LR))
```
