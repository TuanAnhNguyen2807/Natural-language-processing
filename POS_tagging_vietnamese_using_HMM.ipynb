```python
import numpy as np
import pandas as pd
import nltk 
from nltk.util import ngrams
import string
from nltk import word_tokenize
import random
import os
nltk.download('punkt')
```

```python
# loại bỏ các dấu câu kí tự đặc biệt
def removePunctuations(sen):
    temp_l = sen.split()
    i = 0
    j = 0
    for word in temp_l :
        j = 0
        for l in word :
            if l in string.punctuation:
                word = word.replace(l," ")
            j += 1
        temp_l[i] = word.lower()
        i=i+1   
    content = " ".join(temp_l)
    return content
```

```python
# Tạo n-gram
def get_ngrams(text, n ):
    n_grams = ngrams(word_tokenize(text.lower()), n)
    return [ ' '.join(grams) for grams in n_grams]
```

```python
fpaths = []

for f in os.listdir('/content/drive/MyDrive/Kho ngu lieu 10000 cau duoc gan nhan tu loai'):
     fpaths.append('/content/drive/MyDrive/Kho ngu lieu 10000 cau duoc gan nhan tu loai/' + f)
print("All paths:", fpaths)
```

```python
all_sentences = []
for path in fpaths:
    data = pd.read_table(path)
    data_sentences = data.iloc[:,-1].values
    sentences_list = [get_ngrams(data, 1) for data in data_sentences]
    for i in sentences_list:
        all_sentences.append(i)
```

```python
print(len(all_sentences))
print(all_sentences[0:5])
```

```python
temp_ = []
all_sent_split = []
for i, sents in enumerate(all_sentences[0:2000]):
    for j, n in enumerate(sents):
        if n[0] not in string.punctuation:
            temp_.append(n)
        if n  == "/":
            temp = sents[j-1] + sents[j] + sents[j+1]
            temp_.append(temp)
    all_sent_split.append(temp_)
    temp_ = []
```

```python
print(all_sent_split[0:5])
print(len(all_sent_split))
```

```python
# tạo một mảng chứa cái tuple (word, tag)
nltk_data = [] 
temp_ = []
punc = ["...", "“","”","…"]
tags_list = ["n","np","nc","nu","v","a","p","l","m","r","e","c","i","t","b","y","s","x"]
for sent in all_sent_split:
    for i in sent:
        tup = tuple(i.split("/"))
        if len(tup) != 1 and tup[0] not in string.punctuation and tup[0] not in punc and tup[1] in tags_list:
            temp_.append(tup)    
    nltk_data.append(temp_)
    temp_ = []
print(nltk_data)
print(len(nltk_data))
```

```python
# Tạo tập train và test
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(nltk_data, test_size=0.2,random_state=42)
len(train_set),len(test_set)
```

```python
train_tagged_words = [tup for sent in train_set for tup in sent]
len(train_tagged_words)
```

```python
# Tags
T = set([pair[1] for pair in train_tagged_words])
print(len(T))
print(T)
```

```python
def word_given_tag(word, tag, train_bag = train_tagged_words):
    tag_list = [pair for pair in train_bag if pair[1]==tag]
    count_tag = len(tag_list)
    w_given_tag_list = [pair[0] for pair in tag_list if pair[0]==word]
    count_w_given_tag = len(w_given_tag_list)
    
    return (count_w_given_tag, count_tag)
    
```

```python
def t2_given_t1(t2, t1, train_bag = train_tagged_words):
    tags = [pair[1] for pair in train_bag]
    count_t1 = len([t for t in tags if t==t1])
    count_t2_t1 = 0
    for index in range(len(tags)-1):
        if tags[index]==t1 and tags[index+1] == t2:
            count_t2_t1 += 1
    return (count_t2_t1, count_t1)
```

```python
tags_matrix = np.zeros((len(T), len(T)), dtype='float32')
for i, t1 in enumerate(list(T)):
    for j, t2 in enumerate(list(T)): 
        tags_matrix[i, j] = t2_given_t1(t2, t1)[0]/t2_given_t1(t2, t1)[1]
```

```python
tags_df = pd.DataFrame(tags_matrix.T, columns = list(T), index=list(T))
print(tags_df.head(5))
```

```python
def Viterbi(words, train_bag = train_tagged_words):
    state = []
    T = list(set([pair[1] for pair in train_bag]))
    for key, word in enumerate(words):
        #initialise list of probability column for a given observation
        p = [] 
        for tag in T:
            if key == 0:
                transition_p = tags_df.loc[tags_df.index[0], tag]
            else:
                transition_p = tags_df.loc[state[-1], tag]
                
            # compute emission and state probabilities
            emission_p = word_given_tag(words[key], tag)[0]/word_given_tag(words[key], tag)[1]
            state_probability = emission_p * transition_p    
            p.append(state_probability)
            
        pmax = max(p)
        # getting state for which probability is maximum
        state_max = T[p.index(pmax)] 
        state.append(state_max)
    return list(zip(words, state))
```

```python
# list of sents
test_run = test_set

# list of tagged words
test_run_base = [tup for sent in test_run for tup in sent]

# list of untagged words
test_tagged_words = [tup[0] for sent in test_run for tup in sent]
```

```python
# tagging the test sentences
tagged_seq = Viterbi(test_tagged_words)

# accuracy
check = [i for i, j in zip(tagged_seq, test_run_base) if i == j] 

accuracy = len(check)/len(tagged_seq)
print('Viterbi Algorithm Accuracy: ',accuracy*100)
```
